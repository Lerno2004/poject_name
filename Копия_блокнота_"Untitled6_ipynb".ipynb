```python
from datasets import load_dataset
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
ds = load_dataset("Venesa123/livestock-health-disease-ssa-synthetic")
df=ds["train"].to_pandas()
spark=(SparkSession.builder.appName("MYAPP").master("local[*]").getOrCreate())
df=spark.createDataFrame(df)
df.select("agro_ecological_zone", "farm_size_ha").show()
```

```python
df.select(col("agro_ecological_zone"),
    col("farm_size_ha"),
    (col("farm_size_ha") + 500).alias("size_+500"),
    when(col("farm_size_ha") > 30, "Senior").otherwise("Junior").alias("senior_junior")
).show()
```

```python
df.where(col("region_type")=="rural_remote").show()
```

```python
df.filter(col("agro_ecological_zone").like("h%")).show()
```

```python
df.filter(col("livestock_dependency_index").between(0,60)).show()
```

```python
df.groupBy("disease_type").count().show()
df.groupBy("disease_type").sum().show()
df.groupBy("disease_type").min().show()
```

```python
df.groupBy("water_source_reliability").agg(
    count("*").alias("employee_count"),
    avg("household_size").alias("avg_household_size"),
    max("household_size").alias("max_household_size"),
    min("household_size").alias("min_household_size")
).show()
```

```python
employees = spark.createDataFrame([
    (1, "Alice", "HR"),
    (2, "Bob", "Engineering"),
    (3, "Charlie", "Marketing")
], ["id", "name", "dept"])

departments = spark.createDataFrame([
    ("HR", "Human Resources"),
    ("Engineering", "Software Engineering"),
    ("Finance", "Financial Services")
], ["dept", "full_name"])
join_inner=employees.join(departments,"dept","right")
join_inner.show()
```

```python
result = employees.join(
    departments,
    (employees.dept == departments.dept) & (employees.id > 1),
    "inner"
)
result.show()
```

```python
df.createOrReplaceTempView("employees")
result = spark.sql("""
    SELECT department,
           COUNT(*) as employee_count,
           AVG(salary) as avg_salary
    FROM employees
    WHERE age > 25
    GROUP BY department
    ORDER BY avg_salary DESC
""")
```
